```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures, LabelEncoder
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet 
from sklearn.model_selection import train_test_split, cross_val_score,GridSearchCV, cross_val_predict
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, cohen_kappa_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.svm import SVC

import warnings
#warnings.filterwarnings("ignore")
```

```python
ca = pd.read_csv("https://www.dropbox.com/s/s2a1uoiegitupjc/cannabis_full.csv?raw=1")

ca = ca.dropna()
```

I mainly contemplated between "accuracy" and "F1" score and came to the conclusion that F1 is the better option as we are dealing with imbalanced class distribution and it seems tha F1 is better at handling these kinds of models.

Also for F1 to work I had to transform sativa to match value of 1 and indica to match value of 0

# Part 1

```python
uni = ca[(ca['Type'] == "sativa") | (ca['Type'] == "indica") ]

#print(uni)

#print(ca[ca["Type"] == "sativa"].count())
#print(ca[ca["Type"] == "indica"].count())

X = uni.drop(['Type', 'Strain', 'Effects', 'Flavor'], axis = 1)
y = uni['Type'].apply(lambda x: 1 if x == "sativa" else 0)
```

## Q1 LDA

```python
# My research indicates that there is no real parameters that can be changed to tune this model

pipeline = Pipeline(
    [
     ("Model", LinearDiscriminantAnalysis())]
)

scores = cross_val_score(pipeline, X, y, cv=5, scoring = 'f1')

print(scores.mean())
```

## Q2 QDA

```python
# My research indicates that there is no real parameters that can be changed to tune this model

pipeline = Pipeline(
    [
     ("Model", QuadraticDiscriminantAnalysis())]
)

scores = cross_val_score(pipeline, X, y, cv=5, scoring = 'f1')

print(scores.mean())

# This model raised a collinear warning, however, I am not too worried as it did not perform the best (although this might be caused by the collinearity), and since these are user reported subjective values the collinearity is no surprising.
```

## Q3 SVC

```python
pipeline = Pipeline(
    [
     ("LR", SVC(kernel = "linear", C = 0.3))]
)

scores = cross_val_score(pipeline, X, y, cv=5, scoring = 'f1')

print(scores.mean())
```

## Q4 SVM

```python
pipeline = Pipeline(
    [
     ("Model", SVC(kernel = "poly", C = 0.7))]
)

scores = cross_val_score(pipeline, X, y, cv=5, scoring = 'f1')

print(scores.mean())
```

```python
pipeline = Pipeline(
    [
     ("Model", SVC(kernel = "rbf"))]
)

scores = cross_val_score(pipeline, X, y, cv=5, scoring = 'f1')

print(scores.mean())
```

```python
pipeline = Pipeline(
    [
     ("Model", SVC(kernel = "sigmoid"))]
)

scores = cross_val_score(pipeline, X, y, cv=5, scoring = 'f1')

print(scores.mean())
```

```python
ca = pd.read_csv("https://www.dropbox.com/s/s2a1uoiegitupjc/cannabis_full.csv?raw=1")

ca = ca.dropna()

X = ca.drop(['Type', 'Strain', 'Effects', 'Flavor'], axis = 1)
y = ca['Type']
```

## Q1 DT

```python
# This was tuned but I left the code out
pipeline = Pipeline(
    [
     ("DT", DecisionTreeClassifier(max_depth = 3))]
)


scores = cross_val_score(pipeline, X, y, cv=5, scoring = 'f1_weighted')

print(scores.mean())



plt.figure(figsize=(20,10))
plot_tree(DecisionTreeClassifier(max_depth=3).fit(X,y), filled= True, feature_names = X.columns.to_list(), class_names = y.unique().tolist())
plt.show()
```

The best tuned DT model is with a depth of 3. What is notable is that for this purpose we had to use f1_weighted as we have 3 different variables. Most hybrids fell down the very left path, with all there values skewing to the left side of the tree, however, as expected based on the generally understood nature of cannabis, since the first split is based on sleepy or not, sativa mostly lives on the non sleepy side, whereas, indica is on the sleepy side of the tree. It's cool to see it realize in the model as I've only heard people say that this is the case.

## KNN

```python
ct = ColumnTransformer(
  [
    ("standardize", 
    StandardScaler(), 
    ['Rating'])
  ],
  remainder = "passthrough"
).set_output(transform = "pandas")
```

```python
y = ca['Type']
y = LabelEncoder().fit_transform(y)

pipeline = Pipeline(
    [("Preprocessing", ct),
     ("KNN", KNeighborsClassifier())]
)

degrees = {'KNN__n_neighbors': list(range(1,50))}

gscv = GridSearchCV(pipeline, degrees, cv = 5, scoring='f1_weighted')

gscv_fitted = gscv.fit(X, y)

scores = gscv_fitted.cv_results_['mean_test_score']

reshape_scores = scores.reshape(-1,49)

df = pd.DataFrame(reshape_scores)

df
print(df[df.max().idxmax()])
```
